```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

# Load dataset
X, y = load_diabetes(return_X_y=True)

# Gunakan hanya satu fitur (fitur ke-2, yaitu indeks 2)
X = X[:, [2]]

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, shuffle=False)

# Cetak bentuk (shape) dari data
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load dataset
X, y = load_diabetes(return_X_y=True)

# Gunakan hanya fitur ke-2 (biasanya BMI)
X = X[:, [2]]

# Bagi data (tanpa acak), 20 data terakhir jadi test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20, shuffle=False)

# Buat dan latih model regresi linear
regressor = LinearRegression().fit(X_train, y_train)

# Prediksi
y_pred = regressor.predict(X_test)

# Evaluasi
print(f"Mean squared error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of determination (RÂ² score): {r2_score(y_test, y_pred):.2f}")

# Visualisasi hasil prediksi
plt.scatter(X_test, y_test, color="black", label="Data Sebenarnya")
plt.plot(X_test, y_pred, color="blue", linewidth=2, label="Prediksi Regresi")
plt.xlabel("Fitur ke-2 (BMI)")
plt.ylabel("Progression of Diabetes")
plt.title("Regresi Linear - Dataset Diabetes")
plt.legend()
plt.show()

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(10, 5), sharex=True, sharey=True)

ax[0].scatter(X_train, y_train, label="Train data points")
ax[0].plot(
    X_train,
    regressor.predict(X_train),
    linewidth=3,
    color="tab:orange",
    label="Model predictions",
)
ax[0].set(xlabel="Feature", ylabel="Target", title="Train set")
ax[0].legend()

ax[1].scatter(X_test, y_test, label="Test data points")
ax[1].plot(X_test, y_pred, linewidth=3, color="tab:orange", label="Model predictions")
ax[1].set(xlabel="Feature", ylabel="Target", title="Test set")
ax[1].legend()

fig.suptitle("Linear Regression")

plt.show()

```
